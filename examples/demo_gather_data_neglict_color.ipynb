```python
import torch
from nerflab.camera.presets import make_spherical_poses

# 1) Get exactly 20 poses starting at index 100 (from a conceptual total C=1000)
H_wc = make_spherical_poses(
    c=1000, radius=3.0, method="fibonacci",
    offset=500, batch_size=20, device="cpu", dtype=torch.float32
)
print(H_wc.shape)
# # 2) Stream everything in B=20 chunks
# from nerflab.camera.pose_utils import iter_spherical_pose_batches

# for H_wc in iter_spherical_pose_batches(
#         c_total=1000, batch_size=20, radius=3.0, method="fibonacci",
#         device="cpu", dtype=torch.float32):
#     # render/process this batch...
#     print(H_wc.shape)
#     pass


```

```python
# --- camera ------------------------------------------------------------------
# H_wc = np.eye(4, dtype=np.float32)        # simple origin‑facing camera
intr  = Intrinsics(**CFG.intrinsics.__dict__)
cam   = Camera(H_wc, intr,
               t_bounds=(CFG.rays.t_near, CFG.rays.t_far))

# --- sample rays -------------------------------------------------------------
O_w, D_w = cam.get_rays(frame="world")
# O_w, D_w = cam.get_rays_sampled(frame="world", rays_per_pose=100)

# t, delta, pts = stratified_samples_batch(
#     O_w, D_w, t_near = CFG.rays.t_near, t_far = CFG.rays.t_far,
#     N=CFG.rays.N, deterministic=CFG.rays.deterministic
# )
t, delta, pts = cam.sample_along_rays(O_w, D_w)
print("points shape:", pts.shape)
print("t shape:", t.shape)
print("delta shape:", delta.shape)
```
